## Задание 3.2
Найдите всех пассажиров с аномальной стоимостью билета (поле `Fare`). Будем считать стоимость аномальной, если она отличается от среднего более, чем на 5 стандартных отклонений. Выведите соответствующие строки таблицы на экран.

```python
#на самом деле аномально низкой стоимость быть не может т.к. ст.отклонение больше среднего
#т.е. аноальная стоимость была бы отрицательной 
#но для общности добавляем условие |(df.Fare < df.Fare.mean()-5*df.Fare.std())
df.loc[(df.Fare > df.Fare.mean()+5*df.Fare.std())|(df.Fare < df.Fare.mean()-5*df.Fare.std()), 
       ['Name','Fare','Survived']]
```
